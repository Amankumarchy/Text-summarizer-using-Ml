Importing The Dependencies

```python
import nltk
import string
from heapq import nlargest
```

Taking text inputs from the Users

```python
s=input('Enter the text data for summarizing:\n')
```

Generate word cloud image

```python
import wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
print('The length of the text is',len(s))
wc=WordCloud(stopwords=STOPWORDS,background_color='white').generate(s)
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

Importing nltk libraray

```python
import nltk
nltk.download("popular")
```

Keyword Extraction

```python
from rake_nltk import Rake
rk=Rake()
rk.extract_keywords_from_text(s)
extract_keyword=rk.get_ranked_phrases()
extract_keyword
```

  Text Summarization
  

```python
#print(s.count('.'))
#print(string.punctuation)
nopunch=[char for char in s if char not in string.punctuation]
nopunch="".join(nopunch)
process_text=[word for word in nopunch.split() if word.lower() not in nltk.corpus.stopwords.words('english')]
word_freq={}
for word in process_text:
  if word not in word_freq:
    word_freq[word]=1
  else:
    word_freq[word]+=1
max_freq=max(word_freq.values())
for word in word_freq.keys():
  word_freq[word]=word_freq[word]/max_freq
sent_list=nltk.sent_tokenize(s)
sent_score={}
for sent in sent_list:
  for word in nltk.word_tokenize(sent.lower()):
    if word in word_freq.keys():
      if sent not in sent_score.keys():
        sent_score[sent]=word_freq[word]
      else:
        sent_score[sent]+=word_freq[word]
summary_sent=nlargest(3,sent_score,key=sent_score.get)
summary=''.join(summary_sent)
print(len(summary))
print(summary)
```
